```python
data = {}
with open('coutile_computing_results_reputation.csv', 'r') as f:
    for l in f:
        data[l.split('->')[0].strip()] = dict()
        data[l.split('->')[0].strip()]["reputation"] = list(map(float, l.split('->')[1].strip().split(',')))
        
with open('coutile_computing_results_quality.csv', 'r') as f:
    for l in f:
        data[l.split('->')[0].strip()]["quality"] = list(map(float, l.split('->')[1].strip().split(',')))
```

```python
import matplotlib.pyplot as plt
import numpy as np

xx = list(np.arange(0.0, 1.01, 0.01))
```

```python
plt.figure(figsize=(8,5))
plt.ylim((-0.05, 1.1))
for i in (3, 5, 7, 9):
    plt.plot(xx, data[f'(n100_SM3_m10_g{i}_rounds100_priority_binary_1)']['quality'], label=f'g = {i}')
#plt.plot(xx, data[f'(n100_SM3_m10_g3_rounds100_random_nonbinary_1)']['quality'], c='r')
plt.xlabel('Ratio of malicious peers.')
plt.ylabel('Quality of results.')
plt.title('Selection of peers based on reputation (binary task).')
plt.grid()
plt.legend()
#plt.show()
plt.savefig('quality-priority-binary.png')
plt.close()

plt.figure(figsize=(8,5))
for i in (3, 5, 7, 9):
    plt.plot(xx, data[f'(n100_SM3_m10_g{i}_rounds100_random_binary_1)']['quality'], label=f'g = {i}')
plt.xlabel('Ratio of malicious peers.')
plt.ylabel('Quality of results.')
plt.title('Random selection of peers (binary task).')
plt.grid()
plt.legend()
#plt.show()
plt.savefig('quality-random-binary.png')
plt.close()


plt.figure(figsize=(8,5))
plt.ylim((-0.05, 1.1))
for i in (3, 5, 7, 9):
    plt.plot(xx, data[f'(n100_SM3_m10_g{i}_rounds100_priority_nonbinary_1)']['quality'], label=f'g = {i}')
#plt.plot(xx, data[f'(n100_SM3_m10_g3_rounds100_random_nonbinary_1)']['quality'], c='r')
plt.xlabel('Ratio of malicious peers.')
plt.ylabel('Quality of results.')
plt.title('Selection of peers based on reputation (non-binary task).')
plt.grid()
plt.legend()
#plt.show()
plt.savefig('quality-priority-nonbinary.png')
plt.close()


plt.figure(figsize=(8,5))
for i in (3, 5, 7, 9):
    plt.plot(xx, data[f'(n100_SM3_m10_g{i}_rounds100_random_nonbinary_1)']['quality'], label=f'g = {i}')
plt.xlabel('Ratio of malicious peers.')
plt.ylabel('Quality of results.')
plt.title('Random selection of peers (non-binary task).')
plt.grid()
plt.legend()
#plt.show()
plt.savefig('quality-random-nonbinary.png')
plt.close()
```
